```python
#Join shapefile and csv
joincsv = data.join(csv)
joincsv.head()
```

```python
#Remove columns
joincsv.drop(columns=['Type_I','Type_II','Type_III','Type_IV','Type_V'], inplace=True)
#joincsv.drop([1,2,3,4], inplace=True)
joincsv.head()
```

```python
joincsv.plot(column='2016-10-27', legend=True, legend_kwds={"label": "Delta Perimeter (m)", "orientation": "horizontal"})
```

```python
#Saving the date on a variable
days = []
for x in range (11,38):
    days.append(joincsv.columns[x])

print(days)
```

```python
joincsv['DGO_FID'][5]
```

```python
outpath     = "/home/paudisio/Documents/ENS/QGIS/Animation/figures"
vmin, vmax  = 0, 1
n=-1
for day in days:
    n = n+1
    fig, ax = plt.subplots(1,1,figsize=(10,10))
#    joincsv.apply( lambda x: ax.annotate(text=joincsv['DGO_FID'][:], xy=x.geometry.centroid.coords[0], ha='center'), axis=1)
    map = joincsv.plot(ax=ax, column=day, edgecolor= 'black', cmap="Blues" ,vmin=vmin ,vmax=vmax ,legend=True, legend_kwds={"label": "Standard value", "orientation": "horizontal"})
    plt.suptitle('Lhasa river - Delta perimeter')
    map.set_title(f"{day}")
    
    for x, y, label in zip(joincsv.geometry.centroid.x, joincsv.geometry.centroid.y, joincsv['DGO']):
        ax.annotate(label, xy=(x, y), xytext=(3, 3), textcoords="offset points")

    
    filepath = f"{outpath}/{day}.png"
    mapfig = map.get_figure()
    mapfig.savefig(filepath)
    #plt.show()
    plt.close('all')

    print(filepath)
```

```python
from PIL import Image
import glob

frames = []
directory = outpath
folders = os.listdir(directory)

#grab last 4 characters of the file name:
def last_chars(x):
    return(x[:10])

folders = sorted(folders, key = last_chars) 

for i in folders:
    path = outpath+"/"+i
    image_frame = Image.open(path)
    frames.append(image_frame)

frames[0].save(outpath + '/Delta_perimeter.gif', format = 'GIF', append_images=frames[1:], save_all=True, duration=1500, loop=0)
```
